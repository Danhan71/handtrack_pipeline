```python
import os
import pandas as pd
import shutil

checkerboard_cam_order = ["flea", "fly1", "bfs1", "bfs2", "fly2"]

wand_pts_dates = ["220922_grammardircolor1_wandgrid", "220924_grammardircolor1b_wandgrid", "220926_grammardircolor2_wandgrid", "221002_grammardircolor3c_wandgrid", "221015_dircolor2_wandgrid", "221015_dircolor3b_wandgrid", "221015_dirshapecolor1b_wandgrid"]

data_dir = "/data3/hand_track/Pancho"

for date in wand_pts_dates:
    this_dir = f"{data_dir}/{date}/wand/wand_calibration"
    col_list = open(f"{this_dir}/columns.txt").read().splitlines()
    os.remove(f"{this_dir}/{date}-wandPoints.csv")
    wand_pts = [c for c in os.listdir(this_dir) if c.endswith('.csv')][0]
    df = pd.read_csv(f"{this_dir}/{wand_pts}")
    df.columns = col_list
    align_list = []
    for cam in checkerboard_cam_order:
        for col in list(df.columns):
            if cam in col and 'green' in col:
                align_list.append(col)
    for cam in checkerboard_cam_order:
        for col in list(df.columns):
            if cam in col and 'red' in col:
                align_list.append(col)
    df = df[align_list]
    df.to_csv(f"{this_dir}/{date}-wandPoints.csv", index = False, header = False)
    shutil.copy(f"{this_dir}/{date}-wandPoints.csv", f"{data_dir}/220914_wandall/wand/wand_calibration/{date}-wandPoints.csv")
        
```
