```python
def remove_badframes_from_goodframes_all_dan(V, bad_frames):
    indtrial = 0
    ind_group = 0
    dict_cams = V.get_cameras()
    cam_list = [c[1][0] for c in dict_cams.items()]
    mapping_old_new, list_goodframes_oldnum, list_goodframes_newnum \
        = V.goodframes_mapping_new_old_index(ind_group)
    good_frames_dict = {}
    for cam in cam_list:
        good_frames_dict[cam] = [frame for frame in list_goodframes_oldnum \
                                 if frame not in bad_frames[cam]]
    print(good_frames_dict)
def filter_good_frames_dan(V):
    ncams = V.num_cams()
    dict_cams = V.get_cameras()
    cam_list = [c[1][0] for c in dict_cams.items()]
    indtrial = 0
    ind_group = 0

    # Sanity checks
    # 1) One video per cam
    assert V.num_cams()==len(V.DatVideos), f"{V.num_cams()} == {len(V.DatVideos)}"
    assert len(V.DatGroups)==len(V.DatVideos)
    assert len(V.inds_trials())==1
    # 2) Each cam has exact same good frames
    list_good_frames = [dat["good_frames"] for dat in V.DatVideos]
    for list1, list2 in zip(list_good_frames[:-1], list_good_frames[1:]):
        assert list1 == list2

    THRESH = 0.8
    list_part, list_feat = V.dlc_get_list_parts_feats()
    print("HERE")
    bad_frames = {}
    for cam,camname in zip(range(ncams),cam_list):
        bad_frames[camname] = []
        datv = V.wrapper_extract_dat_video(None, cam, indtrial)
        goodframes = datv["good_frames"]
        print("Good frames: (for this cam):", datv["index"])
        print(goodframes)
        for part in list_part:
            # LIKELI
            likeli = V.dlc_data_part_feat(datv["index"], part, "likelihood").values
            for frame in goodframes:
                if likeli[frame]<THRESH:
                    bad_frames[camname].append(frame)
                    
            # NOTE OUT OF BOUNDS 
            out_of_bounds = V.dlc_data_part_feat(datv["index"], part, "out_of_bounds").values
            frames_oob = np.where(out_of_bounds==True)[0].tolist()
            frames_oob = [f for f in frames_oob if f in goodframes] # only keep frames that are in good frames list.
            bad_frames[camname].extend(frames_oob)
    print(f"BAD FRAMES, trial {indtrial}, across all cams (original frame idx:")
    print(bad_frames)

    # Remove any frame that is bad in at least one camera
    print("REMOVING these frames")
    remove_badframes_from_goodframes_all_dan(V, bad_frames)
    
```
