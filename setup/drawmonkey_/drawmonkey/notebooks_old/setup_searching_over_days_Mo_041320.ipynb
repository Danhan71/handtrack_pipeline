```python
# === go over days.
from pythonlib.tools.expttools import getDateList
from tools.preprocess import loadSingleData

dates = getDateList(200219, 200226)
animal = "Mo"
expt = "pilot"
sessions = range(20)

outdict = []
for d in dates:
    for s in sessions:
        # === 1) load filedata
        filedata = loadSingleData(animal, d, expt, s, resave_as_dict=False, load_resaved_data=True, 
                              resave_overwrite=False)
        
        if not filedata is None:
            
            blocks = [getTrialsBlockParams(filedata, t) for t in getIndsTrials(filedata)]
            outdict.append({
                "date":d,
                "sess":s,
                "expt":expt,
                "animal":animal,
                "behtype_trials":[b["behtype"]["type"] for b in blocks],
                "EDGE_FIRST_trials":[b["params_task"]["EDGE_FIRST"] for b in blocks if "EDGE_FIRST" in b.keys()],
                "inkrate_time_trials":[b["params_task"]["inkrate_time"] for b in blocks],
                "taskstages":getMultTrialsTaskStages(filedata),
            })
```
