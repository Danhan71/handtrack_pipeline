```python
## MAKE PLOTS OF GOOD TRIALS FOR NRSA
# Only looked at Pancho for 2/25 and 2/26 

SAVEDIR = "/tmp/figures"

dat = [("Pancho", 200226, 1, 460),
("Pancho", 200226, 1, 77),
("Pancho", 200226, 1, 180),
("Pancho", 200226, 1, 404),
("Pancho", 200226, 1, 568),
("Pancho", 200226, 3, 169),
("Pancho", 200226, 3, 110),
("Pancho", 200226, 3, 5),
("Pancho", 200226, 3, 187),
("Pancho", 200226, 3, 194),
("Pancho", 200226, 3, 200),
("Pancho", 200225, 2, 74),
("Pancho", 200225, 2, 103),
("Pancho", 200225, 2, 442),
("Pancho", 200225, 2, 455),
("Pancho", 200225, 2, 485)]
# dat = [("Pancho", 200226, 1, 180)]

animals = [d[0] for d in dat]
dates = [d[1] for d in dat]
sessions = [d[2] for d in dat]
trials = [d[3] for d in dat]
expts = ["pilot" for _ in range(len(animals))]


for a, d, e, s, t in zip(animals, dates, expts, sessions, trials):

    filedata = loadSingleData(a, d, e, s, resave_as_dict=False, load_resaved_data=True, 
                          resave_overwrite=False)

    sdir = f"{SAVEDIR}/{a}_{d}_{e}_{s}-t{t}"
    import os
    os.makedirs(sdir, exist_ok=True)

    # 1) plot overview
    fig = plotTrialSingleOverview(filedata, t);
    fig.savefig(f"{sdir}/fig1.pdf")

    # 2) plot timelapse
    fig = plotTrialTimelapse(filedata, t, nsteps=6,ncols=4, plot_task_stimulus=False);
    fig.savefig(f"{sdir}/fig2.pdf")
```
