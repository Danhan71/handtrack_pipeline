### [Analysis of all trials together] shapes of gap z jumps

```python
# go thru all trials and collect things
indstrials = getIndsTrials(fd)
dat_all = []
trials_all = []
for trial in indstrials:
    dat,_ = HT.process_data_singletrial(trial, finger_raise_time=0.0)
    if len(dat)>0:
        dat_all.append(dat)
        trials_all.append(trial)
```

```python
# Collect each gap 
list_gaps_betweenstrokes = []
list_gaps_beforedone = []
list_gaps_onset = []
for dat in dat_all:
    gaps_cam = dat["gaps_cam"]
    list_gaps_onset.append(gaps_cam[1])
    list_gaps_betweenstrokes.extend(gaps_cam[2:-1])
    list_gaps_beforedone.extend([gaps_cam[-1]])

    
def plot_gap(gap, ax, plot_vel=False, 
             plot_time_steps=False, val_dim=2):
    #     gap = list_gaps_beforedone[0]
    # align to onset
    t = gap[:,-1]
    t = t-t[0]
    if t[-1]>0.6:
        return
    if t[-1]<0.025:
        return
    if np.any(np.diff(t)>0.04):
        return
        
    z = gap[:,val_dim]
    if z[0]>0.01:
        return
    if z[-1]>0.01:
        return
        
    if plot_vel:
        z = diff5pt(z)
        
    if plot_time_steps:
        t = np.arange(len(z))
        
    # smooth
    z = smooth_it(z)
    
    
    ax.plot(t, z, '-', alpha=0.5)
    ax.axhline(0)
    
    # count zero crossings
    z = gap[:,val_dim]
    z = diff5pt(z)
    z = smooth_it(z)
    zeros = np.where(np.diff(np.sign(z)))[0]
    dur = t[-1] - t[0]
    dist = np.linalg.norm(gap[-1,:2] - gap[0, :2])
    dat = {
        "inds_zero_crossings":zeros,
        "duration":dur,
        "displacement":dist
    }
    return dat




# Plot overlays of all gaps
# fig, ax = plt.subplots(2,2)
DAT = {}
fig, axes = plt.subplots(4,1, sharex=True, sharey=True,figsize=(8, 16))
ax = axes.flatten()[0]
list_dat = []
for gap in list_gaps_onset:
    dat = plot_gap(gap, ax)
    if dat is not None:
        list_dat.append(dat)
DAT["onsets"] = list_dat

ax = axes.flatten()[1]
list_dat = []
for gap in list_gaps_betweenstrokes:
    dat = plot_gap(gap, ax)
    if dat is not None:
        list_dat.append(dat)
DAT["btw_strokes"] = list_dat

# ax.plot(np.linspace(1,100, 10), np.linspace(1,10, 10), 'ok')

# Plot overlays of all gaps
# fig, ax = plt.subplots(2,2)
ax = axes.flatten()[2]
list_dat = []
for gap in list_gaps_beforedone:
    dat = plot_gap(gap, ax)
    if dat is not None:
        list_dat.append(dat)
DAT["offsets"] = list_dat


# Plot overlays of all gaps
# fig, ax = plt.subplots(2,2)
ax = axes.flatten()[3]
list_dat = []
for gap in list_gaps_onset:
    if np.random.rand()>0.5:
        dat = plot_gap(gap, ax)
for gap in list_gaps_betweenstrokes:
    if np.random.rand()>0.5:
        dat = plot_gap(gap, ax)
for gap in list_gaps_beforedone:
    if np.random.rand()>0.5:
        dat = plot_gap(gap, ax)

```

```python
# combine into single dataframe
datflat = []
for k, list_dat in DAT.items():
    for dat in list_dat:
        datflat.append({
            "kind":k,
            "n_zero_crossings":len(dat["inds_zero_crossings"]),
            "duration":dat["duration"],
            "displacement":dat["displacement"]
        })
        

df = pd.DataFrame(datflat)
sns.displot(data=df, x="displacement", y="n_zero_crossings", col="kind")
sns.displot(data=df, x="displacement", y="n_zero_crossings")
# sns.displot(data=df, x="duration", y="n_zero_crossings", hue="kind")
sns.pairplot(data=df, vars=["duration", "n_zero_crossings", "displacement"], kind="hist", height=4)
sns.pairplot(data=df, vars=["duration", "n_zero_crossings", "displacement"], kind="kde", height=4)
sns.pairplot(data=df, vars=["duration", "n_zero_crossings", "displacement"], hue="kind", height=4)
sns.pairplot(data=df, vars=["duration", "n_zero_crossings", "displacement"], hue="kind", kind="kde", height=4)
# sns.pairplot(data=df, vars=["duration", "n_zero_crossings", "displacement"], hue="kind", kind="hist", height=4)
```
