```python
from pythonlib.tools.stroketools import strokesInterpolate2
from moviepy.video.VideoClip import VideoClip

T1 = 1.8
T2 = 12.3

strokes = getTrialsStrokesByPeanuts(fd, trial)

# Get video data
vid_frames, vid_times = extract_video_data(videodat, trial)

# Get params for this trial
if False:
    # instead, use time T1 and T2 hand chosen for cropping.
    if False:
        # only strokes
        duration = strokes[-1][-1,2] # max duration
    else:
        # also vid, then include all vid
        duration = vid_times[-1]
else:
    duration = T2-T1

fig, ax = plt.subplots(1,1)
plotDatStrokes(strokes, ax) # final size of plot
XLIM = ax.get_xlim()
YLIM = ax.get_ylim()

# interpolate so that there is one coordinate per frame.
fsnew = 30 # fps
fsold = fd["params"]["sample_rate"]
strokes = strokesInterpolate2(strokes, ["fsnew", fsnew, fsold])

# concatenate strokes
fig, axes = plt.subplots(1,2, figsize=(6,3))
# plotDatStrokes(strokes, ax)
# ax = plotGetBlankCanvas(fd)

def plotfunc(strokes, ax):
    plotDatStrokes(strokes, ax, alpha=1, each_stroke_separate=True, add_stroke_number=False, 
            force_onsets_same_col_as_strokes=True, naked=True, pcol="k", plotver="onecolor")

def plotfunc_vid(time, ax):
    framenum = np.argmin(np.abs(vid_times-time))
    plotVideoFrame(ax, vid_frames[framenum], crop_xlim = [120, 250], crop_ylim = [135, 280], rotate=180)
    ax.set_yticklabels([])
    ax.set_xticklabels([])
    ax.set_title(None)
    ax.tick_params(axis='both', which='both',length=0)

# method to get frames
def make_frame(t):
    """ t in sec"""
    
    t = t+T1
    
    # 1) Plot strokes
    # clear
    ax = axes.flatten()[0]
    ax.clear()
    
    # only plot strokes up to t
    strokes_this = [strok[strok[:,2]<t, :] for strok in strokes if strok[0,2]<t]
    
    # plotting line
    plotfunc(strokes_this, ax)
    
#     print(len(strokes_this))

#     ax.plot(x, np.sinc(x**2) + np.sin(x + 2 * np.pi / duration * t), lw = 3)
    ax.set_xlim(XLIM[0], XLIM[1])
    ax.set_ylim(YLIM[0], YLIM[1])
    
    # 2) Plot movie
    ax = axes.flatten()[1]
    ax.clear()    
    plotfunc_vid(t, ax)
    
    # returning numpy image
    return mplfig_to_npimage(fig)
 
# creating animation
animation = VideoClip(make_frame, duration = duration)
 
```

```python
# Extract videos
def loadWebcamInfo(filedata, camnum=1):
    import h5py
    import numpy as np
    from tools.utilsh5 import group2dict
    
    animal = fd["params"]["animal"]
    date = fd["params"]["date"]
    expt = fd["params"]["expt"]
    session = fd["params"]["session"]
    fname = findFilename(animal, date, expt, session, doprint=False)[0]

    ################################## LOAD FILE
    print(fname)
    
    with h5py.File(fname, "r") as dat:

        # try:
        # f = h5py.File(fname, "r")
        # except OSError as err:
        #     print("OS error: {0}".format(err))
        #     print(f"Skipping loading of {fname}, NOT PREPROCESSING!!")
        #     return None
    
    
        # data = f["ML"]
        
        # Go thru each trial
        videodat = {}
        for trial in getIndsTrials(filedata):
            this = group2dict(dat["ML"][f"Cam{camnum}_{trial}"])
            fnamefull = this["Filename"]
            inds = [i for i in range(len(fnamefull)) if fnamefull.startswith("\\", i)]
            fnamelast = fnamefull[inds[-1]+1:]

            videodat[trial] = {
                "filename":f"{fd['params']['basedir']}/{animal}/{date}/{fnamelast}",
                "filename_end":fnamelast,
                "frametimes":this["Time"][0]/1000. # sec
            }
                    
        return videodat
        

```

```python
def extract_video_data(videodat, trial):
    # Load the video
    import cv2
    
    # Load the file
    fname = videodat[trial]["filename"]
    assert os.path.exists(fname)
    vidcap = cv2.VideoCapture(fname)
    
    # Extract frames
    count = 0
    success,image = vidcap.read()
    frames = []
    while success:

    #   cv2.imwrite("frame%d.jpg" % count, image)     # save frame as JPEG file      
        success,image = vidcap.read()
#         time = videodat[trial]["frametimes"][count]
        frames.append(image)
        count += 1
    
    # Extract frame times
    times = videodat[trial]["frametimes"]
    assert count == len(videodat[trial]["frametimes"])
    
    return frames, times

frames, times = extract_video_data(videodat, 151)
```

```python
def plotVideoFrame(ax, img, rotate=0, crop_xlim = None, crop_ylim = None):
    """
    - first crops, then rotates
    """
    
    if crop_xlim:
        # applies to y coord. starts numbering from the left
        img = img[:, crop_xlim[0]:crop_xlim[1], :]
    if crop_ylim:
        # applies to x coord. starts numbering from the top
        img = img[crop_ylim[0]:crop_ylim[1], :, :]
    
    if rotate==0:
        pass
    elif rotate==180:
        img = np.flip(img, (0,1))
    else:
        assert False, "code it"
    
    ax.imshow(img)
    

        
```
