```python
from pythonlib.tools.stroketools import strokesVelocity
pcols = getStrokeColors(5)[0]
trials = getIndsTrials(fd)[:200]
trials = [t for t in trials if getTrialsTask(fd, t)["stage"] == "linePlusLv2"]
# trials = [t for t in trials if getTrialsFixationSuccess
W = 7
H = 20
plotkwargs = {
    "marker":"o",
    "alpha":0.8,
    "s":10
}
col="k"
align_by_firsttouch = True
sort_stroknum = True
normalize_strok_lengths = True
fs = fd["params"]["sample_rate"]
plt.figure(figsize=(W, H))
for t in trials:
    strokes = getTrialsStrokesByPeanuts(fd, t)
    

    if not getTrialsFixationSuccess(fd, t):
        continue

    if False:
        _, strokes_speed = strokesVelocity(strokes, fs, lowpass_freq=5)
    #     strokes_speed = getTrialsStrokesSpeed(fd, t, )

        strokes_colors = [s[:,0] for s in strokes_speed]
    else:
        strokes_task = getTrialsTaskAsStrokes(fd, t)
        stroknums_assigned = assignStrokenumFromTask(strokes, strokes_task, sort_stroknum=sort_stroknum)
        strokes_colors = [s for s in stroknums_assigned]
        for i in range(len(strokes_colors)):
            strokes_colors[i] = pcols[strokes_colors[i]]
#         print(strokes_colors)
    
    # - plot each stroke separately
    if align_by_firsttouch:
        t0 = strokes[0][0,2]
    else:
        t0 = 0.
        
    for i, (strok, strokcols) in enumerate(zip(strokes, strokes_colors)):
#         print(strok.shape)
#         print(strokcols.shape)
        
        tvals = strok[:,2] - t0
        if normalize_strok_lengths:
            ontime = (i) + 0.5*i
            tvals = np.linspace(ontime, ontime+1, len(tvals))
            
#         plt.plot(tvals, t*np.ones_like(tvals), color=pcols[i])
#         try:
        plt.scatter(tvals, t*np.ones_like(tvals), c=strokcols, **plotkwargs, 
                   cmap="plasma")
#         except:
#             plt.scatter(tvals, t*np.ones_like(tvals), c="k")
    
    
    if False:
        # flatten strokes into timepoints of samples
        tvals = np.concatenate([s[:,2] for s in strokes], axis=0)
        plt.plot(tvals, t*np.ones_like(tvals), color=col)
```
