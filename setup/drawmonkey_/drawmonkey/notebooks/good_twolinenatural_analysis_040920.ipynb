```python
len_to_take = 0.85
figs_all = []
for output in output_align:
    fs_len = len_to_take*output["first_stroke_median_len"]

    fig, axes = plt.subplots(1,2, figsize=(12,5))
    figs_all.append(fig)
    
    
    # === plot all strokes
    ax = axes[0]
    ax.plot([0, fs_len], [0, 0], '-r')
    for strokes in output["strokes_all"]:
        for s in strokes:
            ax.plot(s[:,1], -s[:,0], '-k', alpha=0.4);

            

    # === plot single strokes
    ax = axes[1]
    do_interp = True
    numskip = 0
    Y =[]
    for strokes in output["strokes_all"]:
        s = strokes[0]

        # -- only keep data before reach the end
        itemindex = np.where(s[:,1]>fs_len)[0]
        if len(itemindex)>0:
            itemindex=itemindex[0]

            # -- do interpolation
            x = np.linspace(0, fs_len, 50)
            y = np.interp(x, s[:itemindex,1], -s[:itemindex,0])

            if do_interp:
                x = len_to_take*x/fs_len
                Y.append(y)
                ax.plot(x, y, '-k', alpha=0.5)
            else:
                # then no inter
                ax.plot(s[:itemindex,1], -s[:itemindex,0], '.k', alpha=0.05);

        else:
            numskip+=1
            pass

    #     if len(itemindex)==0:
    #         print("asdas")
    ax.set_title(f"skipped {numskip}/{len(output['strokes_all'])} trials since did not reach endpoint")
    Y = np.array(Y)
    Ymean = np.mean(Y, axis=0)
    Ystd = np.std(Y, axis=0)
    ax.plot(x, Ymean, '-k', linewidth = 4)    
    ax.plot([0, x[-1]], [0, 0], '-r')
    ax.set_xlabel(f"position along first line, up to {len_to_take} of full")
    ax.set_ylabel("up=towards CCW")
    ax.set_ylim([-60, 60])

for i, fig in enumerate(figs_all):
    fig.savefig(f"{SAVEDIR}/summary_aligned_{i}.pdf")
    
```
