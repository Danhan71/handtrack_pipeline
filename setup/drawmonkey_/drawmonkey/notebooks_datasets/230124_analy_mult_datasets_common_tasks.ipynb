##### Load multiple datasets (daily), then concatenate

```python
from pythonlib.dataset.dataset import load_dataset_daily_helper

dates_dict = {}
dates_dict["Diego"] = [221103]
dates_dict["Pancho"] = [220531, 220603,220614,220616,220624, 220628,220630] # roughly, these are all his chars during recordings

list_D = []
list_animal = []
list_date = []
for animal, dates in dates_dict.items():
    for date in dates:
        list_animal.append(animal)
        list_date.append(date)
        D = load_dataset_daily_helper(animal, date)
        list_D.append(D)
```

```python
# Concatenate
from pythonlib.dataset.analy_dlist import concatDatasets, find_common_tasks
```

```python
# find tasks that are present for each animal.
Dall = concatDatasets(list_D)
```

```python
# Sanity check - any tasks have same fixe set, but different hash? This means
# slight different in numeerical values for pts.
Dall._check_character_fixed_los_not_multiple_hashes()

```

```python
# For each unique char, print example drawings across each animal-date combination
# from pythonlib.tools.pandastools import append_col_with_grp_index
# Dall.Dat = append_col_with_grp_index(Dall.Dat, ["animal", "epoch"], new_col_name="animal-epoch", strings_compact=True)
Dall.grouping_append_col(["animal", "epoch"], new_col_name="animal-epoch", strings_compact=True)
```

```python
##### Saving 

SDIR_MAIN = Dall.make_savedir_for_analysis_figures("mult_datasets_common_tasks")
from pythonlib.tools.expttools import writeDictToYaml
writeDictToYaml(dates_dict, f"{SDIR_MAIN}/dates_dict.yaml")


# get all fixed chars
df = Dall.Dat
df = df[(df["task_kind"]=="character") & (df["random_task"]==False) & (df["supervision_stage_concise"]=="off|0||0")]

# chars that have at least 1 trial for both animals
from pythonlib.tools.pandastools import prune_min_ntrials_across_higher_levels
df = prune_min_ntrials_across_higher_levels(df, "animal", "character", 1)

list_char_unique = sorted(df[(df["task_kind"]=="character") & (df["random_task"]==False)]["character"].unique().tolist())

print("Found this many unique chars: ", len(list_char_unique))
for char in list_char_unique:
    print(char)


from pythonlib.tools.pandastools import grouping_get_inner_items, grouping_print_n_samples

grouping_print_n_samples(df, ["character", "animal", "epoch"])
groupdict = grouping_get_inner_items(df, "character", "animal-epoch")
writeDictToYaml(groupdict, f"{SDIR_MAIN}/grouping_character_animal_epoch.yaml")


from pythonlib.dataset.dataset_analy.summary import _plot_summary_drawing_examplegrid

# Check if plot is large. if so, split into multiple plots.
subfolder = "test"
_plot_summary_drawing_examplegrid(df, SDIR_MAIN, subfolder=subfolder, 
                 yaxis_ver="animal-epoch", how_to_split_files = "taskgroup")


# For each individual task, plot all trials
from pythonlib.dataset.dataset_analy.summary import _plot_summary_drawing_eachtrial

# Check if plot is large. if so, split into multiple plots.
SAVEDIR_FIGS = "/tmp"
subfolder = "test"
_plot_summary_drawing_eachtrial(df, SDIR_MAIN, subfolder=subfolder, 
                 row_variable="animal-epoch")
```
