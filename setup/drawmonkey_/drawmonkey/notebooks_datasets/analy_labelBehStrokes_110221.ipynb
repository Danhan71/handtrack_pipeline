```python
# Pull out the task strokes that fall under a threshold for the max match
def _plot_cases_with_low_simscores(D, threshmax = 0.5, threshmin = 0.2, axis = 0, n=20, method="uniform"):
    list_max = []
    list_inds = []
    for i in range(len(D.Dat)):
        Beh = D.Dat.iloc[i]["BehClass"]
        smat = Beh.Alignsim_simmat_sorted
        maxes = np.max(smat, axis=axis)
        for j, m in enumerate(maxes):
            list_max.append(m)
            list_inds.append((i,j))
    sims, indices = D.analy_subsample_trials_for_example(list_max, list_inds, 
                                                         score_range=[threshmin, threshmax], 
                                                         method=method, n=n)

    # - plot
    indtrials = [i[0] for i in indices]
    indstrokes = [i[1] for i in indices]
    list_strokes_beh = D.Dat["strokes_beh"].values[indtrials]
    list_beh = D.behclass_extract(indtrials)
    list_strokes_task = [b.extract_strokes("task_after_alignsim") for b in list_beh]
    list_strokes_task_orig = D.Dat["strokes_task"].values[indtrials]
    
    fig1, axes1 = D.plotMultStrokesByOrder(list_strokes_beh, titles=indtrials, plotkwargs={"number_from_zero":True})
    fig2, axes2 = D.plotMultStrokesByOrder(list_strokes_task, titles=indstrokes, plotkwargs={"number_from_zero":True})
    fig3, axes3 = D.plotMultStrokes(list_strokes_task_orig, titles=indstrokes, is_task=True)
    return fig1, fig2, fig3, indtrials, sims
```

```python
#### Plot, asking which task strokes are bad
threshmax = 0.5
for threshmin in [0.0, 0.2]:
    axis = 0
    n=20
    fig1, fig2, fig3 = _plot_cases_with_low_simscores(D, threshmax = threshmax, threshmin = threshmin, axis = axis, n=n)

    # save
    path = f"{SDIR}/example_missed_taskstrokes-threshmin_{threshmin}"
    saveMultToPDF(path, [fig1, fig2, fig3])


#### Plot, asking which beh strokes are bad
threshmax = 0.5
threshmin = 0.
axis = 1
n=10
_plot_cases_with_low_simscores(D, threshmax = threshmax, threshmin = threshmin, axis = axis, n=n)
```
