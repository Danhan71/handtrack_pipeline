```python
## Collecting wildcard matches

# EXTRACT AND SAVE ALL CONCRETE STRING MATCHES
# For each beh trial, extract locations of each motif
LIST_MOTIFS = [
    ("repeat", {"shapekey":"shape",
                 "shape":"circle",
                 "nmin":2,
                 "nmax":4}),
    ("repeat", {"shapekey":"shape",
                 "shape":"line",
                 "nmin":2,
                 "nmax":4}),
    ("lolli", {})
]

# EXTRACT AND SAVE ALL CONCRETE STRING MATCHES
# For each beh trial, extract locations of each motif
DAT = []
for (motifkind, motif_params) in LIST_MOTIFS:
    print(motifkind, motif_params)

    for i in range(len(D.Dat)):
        Beh = D.Dat.iloc[i]["BehClass"]
        matches = Beh.alignsim_find_motif_in_beh_wildcard(motifkind, motif_params)
        
        # Get matches as list
        dict_matches = matches
        matches = [vv for v in matches.values() for vv in v if len(v)>0]
        
        # store
        DAT.append({
            "indtrial":i,
            "motif_kind":motifkind,
            "motif_params":motif_params,
            "matches":matches,
            "dict_matches":dict_matches})
        
        # store motif_params too
        for k, v in motif_params.items():
            if isinstance(v, dict):
                # go deeper
                for kk, vv in v.items():
                    assert kk not in DAT[-1].keys()
                    DAT[-1][kk] = vv
            else:
                assert k not in DAT[-1].keys()
                DAT[-1][k] = v
        
        # how many matches
        DAT[-1]["n_matches"] = len(matches)
        DAT[-1]["any_match"] = len(matches)>0


import pandas
DAT_DF = pd.DataFrame(DAT)
DAT_DF
```

```python
##### insert back in to Dataset
# Clear all columns with nmatches
for c in D.Dat.columns:
    if "n_matches" in c:
        try:
            del D.Dat[c]
        except:
            pass
        
# Collect
D.Dat["indtrial"] = D.Dat.index
list_col_names = []
for (motifkind, motifparams) in LIST_MOTIFS:
    
    if motifkind=="repeat":
        F = {"motif_kind":motifkind, "shape":motifparams["shape"]}
    elif motifkind=="lolli":
        F = {"motif_kind":motifkind}
    else:
        assert False
    
    df = filterPandas(DAT_DF, F)
    assert len(df)==len(D.Dat), "should have one value for each trial"
#     display(df)
    
    # get its shorthand name
    name = Beh.motif_shorthand_name(motifkind, motifparams)
    name = f"n_matches-{name}"
    
    # Replace with updated name
    df[name] = df["n_matches"]
    del df["n_matches"]
    
    # Merge
    # - only keep certain columns of df
    D.Dat = pd.merge(D.Dat, df[["indtrial", name]], how="outer", on="indtrial")
    # replace nans with 0
    D.Dat = D.Dat.fillna({name:0})
    
    print(name)
    list_col_names.append(name)
    
```
