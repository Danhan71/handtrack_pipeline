```python
# Distribution of scores

ncols = 2
nrows = 3
SIZE = 3
fig, axes = plt.subplots(nrows, ncols, sharex=True, sharey=False, figsize=(ncols*SIZE*1.25, nrows*SIZE))

ax = axes.flatten()[0]
ax.set_title("all strokes")
sns.histplot(data=DS.Dat, x="clust_sim_max", ax=ax)
ax.axvline(THRESH, color="r")

ax = axes.flatten()[1]
ax.set_title("all chars (mean score)")
dfthis = Dthis.Dat
sns.histplot(data=dfthis, x="strokes_clust_score", ax=ax)
ax.axvline(THRESH, color="r")

ax = axes.flatten()[2]
ax.set_title("all chars (min score)")
dfthis = Dthis.Dat
sns.histplot(data=dfthis, x="strokes_clust_score_min", ax=ax)
ax.axvline(THRESH, color="r")

# fraction of trials that have all strokes above some threshold for score
list_thresh = np.linspace(0, 1, 21)
list_frac = []
for thresh in list_thresh:
    n = sum(Dthis.Dat["strokes_clust_score_min"]>=thresh)
    ntot = len(Dthis.Dat)
    list_frac.append(n/ntot)

ax = axes.flatten()[3]
ax.plot(list_thresh, list_frac, "-ok")
ax.axvline(THRESH, color="r")
ax.set_ylabel("frac trials w/ all strokes > thresh")
ax.set_xlabel("thresh (strokes clust score)")

# put horizontal line at the single thresh value
n = sum(Dthis.Dat["strokes_clust_score_min"]>=THRESH)
ntot = len(Dthis.Dat)
frac = n/ntot
ax.axhline(frac, color="r")
ax.set_title(f"Frac at score {THRESH} = {frac:.2f}")

# Frac trials with mean score above thresh.
list_frac = []
for thresh in list_thresh:
    n = sum(Dthis.Dat["strokes_clust_score"]>=thresh)
    ntot = len(Dthis.Dat)
    list_frac.append(n/ntot)

ax = axes.flatten()[4]
ax.plot(list_thresh, list_frac, "-ok")
ax.axvline(THRESH, color="r")
ax.set_ylabel("frac trials w/ all strokes > thresh")
ax.set_xlabel("thresh (strokes clust score)")

# put horizontal line at the single thresh value
n = sum(Dthis.Dat["strokes_clust_score"]>=THRESH)
ntot = len(Dthis.Dat)
frac = n/ntot
ax.axhline(frac, color="r")
ax.set_title(f"Frac at score {THRESH} = {frac:.2f}")
```
