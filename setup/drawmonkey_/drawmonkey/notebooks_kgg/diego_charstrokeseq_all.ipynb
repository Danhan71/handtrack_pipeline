## 2. single day: 220929

#### Load data

```python
# load data
from pythonlib.dataset.dataset_preprocess.general import get_rulelist

expt = "charstrokeseq2"
animal = "Diego"
rulelist = ["220929"] # rulelist = get_rulelist(animal, expt)
D = Dataset([])

D.load_dataset_helper(animal, expt, ver="mult", rule=rulelist)

#from pythonlib.dataset.dataset_preprocess.general import preprocessDat
#D._analy_preprocess_done = False
#D, GROUPING, GROUPING_LEVELS, FEATURE_NAMES, SCORE_COL_NAMES = preprocessDat(D, expt)
```

####  Plot: Nbeh_strokes v. Ntask_strokes (matplotlib)

```python
# select all test tasks
n_data = D.Dat[D.Dat['monkey_train_or_test']=='test']

# for this specific expt, there are char_star, char tasks
# we just want char_star, which comes from sets 15-18
n_data = n_data[n_data['task_stagecategory'].str.contains("ss-1[5-8]")]

# plot num_strokes vs len(strokes_task)
n_data['beh_strokes_count'] = n_data['strokes_beh'].apply(len)
n_data['task_strokes_count'] = n_data['strokes_task'].apply(len)

plt.scatter(n_data['task_strokes_count'],n_data['beh_strokes_count'],alpha=0.5,marker="o")
```

####  Plot: Nbeh_strokes v. Ntask_strokes (seaborn)

```python
import seaborn as sns

sns.catplot(data=n_data,x='task_strokes_count',y='beh_strokes_count',kind='violin')
fig,axes=plt.subplots(1,1)
sns.regplot(data=n_data,x='task_strokes_count',y='beh_strokes_count',ax=axes)
axes.set_xlim([1,6])
axes.set_ylim([1,6])
```
